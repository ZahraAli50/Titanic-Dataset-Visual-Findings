```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('train.csv')
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.info()
```

```python
df['Sex'].value_counts()
```

```python
df_numeric=df.select_dtypes(include=['float', 'int'])
df_numeric.head()
```

```python
sns.heatmap(df_numeric)
```

```python
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm')
```

```python
sns.pairplot(df_numeric, hue='Survived')
```

```python
sns.countplot(x='Sex', data=df, color='skyblue')
plt.xlabel('Gender')
plt.show()
```

```python
sns.countplot(x='Pclass', data=df, color='purple')
plt.xlabel('Passenger Class')
plt.show()
```

```python
sns.countplot(x='Embarked', data=df, color='pink')
plt.xlabel('Emabarkation')
plt.show()
```

```python
sns.histplot(df['Age'], kde=True)
plt.show()
```

```python
sns.histplot(df['Fare'], kde=True, color='brown')
plt.show()
```

```python
pd.crosstab(df['Sex'], df['Survived']).plot(kind='bar', stacked=True)
plt.show()
```

```python
sns.boxplot(x=df['Pclass'], y=df['Age'], color='lightgreen')
plt.show()
```

```python
plt.figure(figsize=(18,20))
for i  in range(len(df_numeric.columns)):
  plt.subplot(3,3,i+1)
  sns.boxplot(x=df[df_numeric.columns[i]], color='lightgreen')
plt.show()
```

```python
df['age_bin']=pd.cut(df['Age'], bins=[0,10,20,30,40,50,60])
sns.barplot(x='age_bin', y='Survived', data=df, color='violet')
plt.show()
```
